```python
import urllib
import pandas as pd
import json
```

```python
exoclock_planets = json.loads(urllib.request.urlopen('https://www.exoclock.space/database/planets_json').read())
```

```python
df_exoclock_planets = pd.DataFrame.from_dict(exoclock_planets,orient='index')
```

```python
df_exoclock_planets
```

```python
df_exoclock_planets.to_excel("ExoClock_planets_list_20260122.xlsx",sheet_name="ExoClock_planets",index=False)
```
